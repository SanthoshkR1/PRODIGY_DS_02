```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.fillna({"Cabin": "Unknown", "Age": data["Age"].mean()}, inplace=True)
```

```python
data["Cabin"] = data["Cabin"].fillna("Unknown")
data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data["Age"],kde=True)
plt.title("Age Distributed")
plt.xlabel("Age")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(data=data,x="Sex",hue="Sex")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(titeles="Survived",loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(data=data,x="Sex",hue="Sex")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(titele="Survived",loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(data=data,x="Sex", hue="Sex")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=data,x="Age",y="Fare",hue="Age")
plt.title("Scatter plot of Age and Fare  ")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
